```python
import matplotlib.pyplot as plt
import numpy as np

# 创建示例数据
categories = ['Return', 'CPI', 'RV', 'Volume', 'Money']
values = [0.173908, 0.4, 0.011045, 11023379.17, 1.167773e+08]  # 使用最新一期数据

# 数据标准化处理
values_normalized = [v/max(values) for v in values]

# 计算角度
angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False)

# 闭合数据
values_normalized = np.concatenate((values_normalized, [values_normalized[0]]))
angles = np.concatenate((angles, [angles[0]]))

# 创建图形
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# 绘制雷达图
ax.plot(angles, values_normalized)
ax.fill(angles, values_normalized, alpha=0.25)

# 设置刻度标签
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# 添加标题
plt.title('Financial Indicators Radar Chart')

# 显示图形
plt.show()
```

```python
import pandas as pd

# 创建相关系数矩阵
correlation_matrix = pd.DataFrame({
    'Return': values_normalized[:-1],
    'CPI': values_normalized[:-1], 
    'RV': values_normalized[:-1],
    'Volume': values_normalized[:-1],
    'Money': values_normalized[:-1]
}, index=categories)

# 创建热力图
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix.corr(), cmap='coolwarm', aspect='auto')
plt.colorbar()

# 添加标签
plt.xticks(range(len(categories)), categories, rotation=45)
plt.yticks(range(len(categories)), categories)

# 添加相关系数值
for i in range(len(categories)):
    for j in range(len(categories)):
        plt.text(j, i, f'{correlation_matrix.corr().iloc[i, j]:.2f}', 
                ha='center', va='center')

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

```

step2；引入数据

（一）引入1990-2024年日收益率数据

```python
data=pd.read_csv('000001.csv')                              #读取名为'000001.csv'的CSV文件
data['Day'] = pd.to_datetime(data['Day'],format='%Y/%m/%d') # 将'Day'列转换为日期类型，指定日期格式为年/月/日
data.set_index('Day',inplace=True)                          # 将'Day'列设置为DataFrame的索引
data['Close']=pd.to_numeric(data['Close'])                  # 将'Close'列中的数据转换为数值类型
data['preclose']=data['Close'].shift(1)                     # 创建'preclose'列，将'Close'列的数据向下移动一行
data['Return']=(data['Close']/data['preclose'])-1           # 计算每日收益率
data
```

（二）采取月度收益率数据

```python
# 从 'data' DataFrame 中提取时间范围在 1995 年 1 月到 2024 年 9 月之间的数据，并创建其副本 'data_new'。
data_new = data['1995-01':'2024-09'].copy() 
# 将 'data_new' 以月为频率重新采样，提取 'Return' 列，并计算每月收益率。                                              
Month_data = data_new.resample('M')['Return'].apply(lambda x: (1+x).prod()-1).to_frame()
Month_data
```

（三）提取1987年-2024年cpi数据

```python
infation = pd.read_csv('inflation.csv')                                    #读取 'inflation.csv' 文件，创建一个 DataFrame 
infation['month'] = pd.to_datetime(infation['month'], format='%Y/%m/%d')   #将 'month' 列的数据类型转换为日期格式
infation.set_index('month', inplace=True)                                  #将 'month' 列设为 DataFrame 的索引
infation
```

（四）计算月度市场收益波动率

```python
market_variance = data_new.resample('M').apply({
    'Return':
    lambda x: sum(x**2)
})
market_variance.reset_index(inplace=True)
market_variance.rename(columns={'Return': 'RV','Day': 'month'}, inplace=True)
market_variance.set_index('month', inplace=True)
market_variance
```

（五）合并以上数据

```python
reg_data = pd.merge(Month_data, infation, left_index=True, right_index=True)
reg_data = pd.merge(reg_data, market_variance, left_index=True, right_index=True)
reg_data
```

step3：预测中国股票市场收益率与波动率之间关系

No.1使用月度数据

（一）画图

```python
reg_data_plot = reg_data['2000':].copy()

fig,ax1 = plt.subplots(figsize=(8,4))
ax1.plot(reg_data_plot['Return'],
        color='red',
        linestyle='-',
        label='Return')
ax1.set_ylabel('Return',color = 'red')

ax2 = ax1.twinx()
ax2.plot(reg_data_plot['RV'].shift(1),
        color='blue',
        linestyle='-',
        label='Realized Variance')

ax2.set_ylabel('Realized Variance',color = 'blue')

plt.title("China's Stock Market Return and Realized Variance")

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.show();
```

（二）OLS Regression最小二乘估计

```python
reg_data['lRV'] = reg_data['RV'].shift(1)

model_rv = smf.ols('Return ~ lRV', data= reg_data['2000':]).fit(
    cov_type='HAC',
    cov_kwds={'maxlags': 6}
)
print(model_rv.summary())
```

```python
reg_data['lcpi'] = reg_data['cpi'].shift(2)

model_rv_cpi = smf.ols('Return ~ lRV + lcpi', data= reg_data['2000':]).fit(
    cov_type='HAC', 
    cov_kwds={'maxlags': 6}
)
print(model_rv_cpi.summary())
```

（一）选取范围
本研究选择了2006年3月至2024年10月的十年期国债收益率和中国股票市场的收益率数据，十年期国债收益率通常被视为长期无风险收益率的代表，对股票市场有着深远影响，选取这段时间的数据，涵盖了中国经济经历的多个重要发展阶段，如2008年全球金融危机、2015年股市震荡、2020年新冠疫情等关键事件，从而可以全面反映国债收益率波动对股市的影响，同时，这段时间的样本数据量较大，这段时间涵盖了不同经济周期和政策环境，能够为研究提供足够的统计样本，有助于提高分析结果的稳健性，捕捉经济政策、利率变化对股市的长期影响，提供更加全面、准确的研究结论。

（二）数据来源
本研究中的十年期国债收益率数据来自中国财政部发布的收益率曲线数据，来源于国家债券市场的交易记录和市场行为，本数据具有时效性，可以满足研究中对于数据准确性和完整性的要求，同时数据公开透明且具有一定的权威性和准确性，可以可靠地反映十年期国债收益率市场的真实情况。

（三）数据特点
本研究数据的特点在于时间跨度长、波动规律明确，并具有对宏观经济和股票市场的高度敏感性，适合用于研究宏观经济变量对股票市场的长期影响：
（1）时间跨度长。2006年3月至2024年10月的十年期国债收益率数据覆盖了中国资本市场从快速增长到多次波动的完整周期，不仅展现了短期波动，也揭示了长期趋势。
（2）波动规律明确。年期国债收益率作为相对稳定的无风险利率基准，其波动率通常低于短期利率，便于分析其对股票市场的持续影响。

（四）数据优势

二、理论分析

（一）影响机制

（1）正向影响（即十年期国债收益率越高，股票收益率越高）
path1：经济增长预期
   在经济繁荣期，国债收益率和股票收益率往往同时上升，体现经济增长驱动的正相关性。国债收益率反映市场对经济的预期，十年期国债收益率上升通常反映市场对经济增长的乐观预期，在经济扩张期，投资者预计企业盈利会改善，进而推动股票市场上涨。
path2：投资者情绪/风险偏好提升
   当经济前景看好时，投资者的风险偏好往往会提升，风险偏好提升引发资金从避险资产（如国债）流向高风险资产（如股票）。债市资金的流出导致国债价格下跌、收益率上升，同时流入股票市场的资金推高股市价格，抬高股票收益率。从而推动十年期国债收益率和股票收益率之间出现正相关关系。
path3：通胀预期与名义收益率同步提升
   如果十年期国债收益率上升的主要驱动力是通胀预期上升，股票作为通胀对冲资产，可能表现较好。这是因为通胀通常伴随着商品价格上涨和企业收入增加，尤其是资源类和周期性行业。企业盈利能力增强会提高市场对股票的估值，同时通胀压力下的国债收益率走高。例如，在全球经济复苏期间，商品价格上涨、通胀上升，股票市场（尤其是能源、原材料板块）也可能表现强劲。
path4:资产配置效应
   在经济上行周期，资金相对倾向于风险资产，投资者再平衡资产配置，债券市场资金可能流入股票市场，导致股票价格上涨，从而使得股市收益提升。而资金从债券市场流出后，债券价格下降，国债收益率上升。




















（2）负向影响（即十年期国债收益率越高，股票收益率越低）
path1：无风险利率基准效应
十年期国债收益率作为无风险利率的长期基准，直接影响股票的折现率，收益率上升时，折现率提高，股票市场的现值降低，收益率受到抑制；反之，收益率下降则支持股市收益。
path2：投资替代效应
当国债收益率较高时，其吸引力上升，投资者倾向于转向债市，减少股票投资需求，压低股市收益
path3：市场情绪效应
国债收益率的变化影响市场情绪。收益率大幅上升可能使投资者对未来经济增长过热或通胀上升的担忧加剧，从而降低股市需求；收益率下降则反映出市场对经济悲观的预期，抑制股市情绪。
path4：风险溢价调整
国债收益率上升会提升投资者对股票的风险溢价要求，导致股市估值下降；当收益率下降时，风险溢价要求降低，股票市场估值水平相对提升。

（4）年期国债收益率的相对稳定性使其成为股票市场收益率的理想参照指标。作为无风险收益率基准，国债收益率的波动性虽低，但与股票市场收益率之间存在显著的传导关系，国债市场在货币政策、经济周期的影响下，往往先于股票市场产生反应，因此分析国债收益率的变化可以为股票市场收益率的波动提供先导信号，这种债券与股票市场之间的关系使得国债收益率数据成为研究股票市场收益率的重要参考指标，有助于更全面地理解宏观经济变量对股票市场的影响。
